Get your `client_id` and `client_secret` to generate your __token__ access

```python
# Generate token with a POST request

client_id = 'f45b96d2a2ce4206a41475966e754b70' # CLIENT ID
client_secret = 'f6a2b132ff9d4f1f8b1036f2f8b8a01b' # CLIENT SECRET
auth_url = 'https://accounts.spotify.com/api/token'
```

```python
auth_response = requests.post(auth_url, {'grant_type': 'client_credentials',
                                         'client_id': client_id,
                                         'client_secret': client_secret}).json()
```

```python
access_token = auth_response['access_token']
auth_response
```

### Set your main variables!!!

Set the `base_uri` (i.e.: end-point), parameters and `headers` for your __GET__ operations

```python
# Base end-point construction

base_url = 'https://api.spotify.com/v1/'
resource = 'artists/'

header_info = {'Authorization': 'Bearer {token}'.format(token=access_token)}
```

```python
# Here you can complete your endpoint URI

full_endpoint = base_url + resource + "2d0hyoQ5ynDBnkvAbJKORj/" + "related-artists"
full_endpoint
```

```python
# Start building your playlist!!!

response = requests.get(full_endpoint, headers=header_info)
related = response.json()['artists']
related
#related_info = related['artists']
#related_info
```

```python


related_artist = [i['id'] for i in related]
len(related_artist)
#related_artist
```

```python
response = []
for i in related_artist:
    response.append(requests.get(base_url + resource + i + "/top-tracks?country=ES", headers=header_info).json())
response[0]['tracks']
```

```python
top_list = []
for i in response:
    
    for j in range(10):
        top_list.append(i['tracks'][j]['name'])


top_list_artists = []
for i in response:
    for j in range(10):
        top_list_artists.append(i['tracks'][j]['album']['name'])


popularity = []
for i in response:
    for j in range(10):
        popularity.append(i['tracks'][j]['popularity'])

release_date = []
for i in response:
    
    for j in range(10):
        release_date.append(i['tracks'][j]['album']['release_date'])
```

```python
df = pd.DataFrame({'Songs': top_list, 'Artist': top_list_artists, 'Popularity': popularity, 'Release Date': release_date})
df
```
